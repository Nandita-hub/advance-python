```python
class InvalidStudentID(Exception):
    pass

class EmptyGradeError(Exception):
    pass

students = {}

try:
    sid = input("Enter Student ID: ")

    if sid in students:
        raise InvalidStudentID("Student ID already exists")

    grade = input("Enter Grade: ")

    if grade == "":
        raise EmptyGradeError("Grade cannot be empty")

    grade = int(grade)
    students[sid] = grade

    print("Student added successfully")

except InvalidStudentID as e:
    print(e)

except EmptyGradeError as e:
    print(e)

except ValueError:
    print("Grade must be a number")

print(students)
```

```python
class DuplicateContactError(Exception):
    pass

class InvalidPhoneError(Exception):
    pass

contacts = {}

try:
    name = input("Enter Name: ")
    phone = input("Enter Phone Number: ")

    if name in contacts:
        raise DuplicateContactError("Contact already exists")

    if not phone.isdigit() or len(phone) != 10:
        raise InvalidPhoneError("Invalid phone number")

    contacts[name] = phone
    print("Contact saved successfully")

except DuplicateContactError as e:
    print(e)

except InvalidPhoneError as e:
    print(e)

print(contacts)
```

```python
class OverdraftError(Exception):
    pass

class InvalidAccountError(Exception):
    pass

accounts = {101: 5000, 102: 3000}

try:
    acc = int(input("Enter Account Number: "))
    amount = int(input("Enter Amount: "))

    if acc not in accounts:
        raise InvalidAccountError("Account number not valid")

    if accounts[acc] < amount:
        raise OverdraftError("Insufficient balance")

    accounts[acc] -= amount
    print("Transaction successful")

except InvalidAccountError as e:
    print(e)

except OverdraftError as e:
    print(e)

print(accounts)
```

```python
class OutOfStockError(Exception):
    pass

class InvalidCouponError(Exception):
    pass

stock = {"laptop": 3, "phone": 5}

try:
    product = input("Enter product name: ")
    qty = int(input("Enter quantity: "))
    coupon = input("Enter coupon code: ")

    if stock[product] < qty:
        raise OutOfStockError("Product out of stock")

    if coupon != "SAVE10":
        raise InvalidCouponError("Invalid coupon code")

    stock[product] -= qty
    print("Order placed successfully")

except OutOfStockError as e:
    print(e)

except InvalidCouponError as e:
    print(e)
```

```python
class OutOfStockError(Exception):
    pass

class InvalidProductIDError(Exception):
    pass

products = {1: 10, 2: 0}

try:
    pid = int(input("Enter Product ID: "))

    if pid not in products:
        raise InvalidProductIDError("Invalid product ID")

    if products[pid] == 0:
        raise OutOfStockError("Product not available")

    print("Product purchased")

except InvalidProductIDError as e:
    print(e)

except OutOfStockError as e:
    print(e)
```
