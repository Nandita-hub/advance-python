```python
# Library Management System

library = {}   # dictionary to store books

while True:
    print("\n----- LIBRARY MENU -----")
    print("1. Add Book")
    print("2. Issue Book")
    print("3. Return Book")
    print("4. Show All Books")
    print("5. Exit")

    choice = int(input("Enter your choice: "))

    # ADD BOOK
    if choice == 1:
        book_id = input("Enter Book ID: ")
        book_name = input("Enter Book Name: ")
        library[book_id] = {"name": book_name, "status": "Available"}
        print("Book added successfully!")

    # ISSUE BOOK
    elif choice == 2:
        book_id = input("Enter Book ID to issue: ")
        if book_id in library:
            if library[book_id]["status"] == "Available":
                library[book_id]["status"] = "Issued"
                print("Book issued successfully!")
            else:
                print("Book is already issued.")
        else:
            print("Book not found.")

    # RETURN BOOK
    elif choice == 3:
        book_id = input("Enter Book ID to return: ")
        if book_id in library:
            library[book_id]["status"] = "Available"
            print("Book returned successfully!")
        else:
            print("Book not found.")

    # SHOW ALL BOOKS
    elif choice == 4:
        if len(library) == 0:
            print("No books in library.")
        else:
            print("\nBook ID | Book Name | Status")
            for i in library:
                print(i, "|", library[i]["name"], "|", library[i]["status"])

    # EXIT
    elif choice == 5:
        print("Thank you for using Library Management System.")
        break

    else:
        print("Invalid choice. Try again.")
```

```python
# Traffic Light Simulation

while True:
    print("\n----- TRAFFIC LIGHT MENU -----")
    print("1. Red Light")
    print("2. Yellow Light")
    print("3. Green Light")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        print(" RED LIGHT")
        print("Stop! Vehicles must wait.")

    elif choice == 2:
        print(" YELLOW LIGHT")
        print("Get Ready! Signal is about to change.")

    elif choice == 3:
        print(" GREEN LIGHT")
        print("Go! Vehicles can move.")

    elif choice == 4:
        print("Traffic Light Simulation Ended.")
        break

    else:
        print("Invalid choice. Please select again.")
```

```python
# Password Strength Checker

password = input("Enter your password: ")

upper = 0
lower = 0
digit = 0
special = 0

if len(password) >= 8:
    for ch in password:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1
        elif ch.isdigit():
            digit += 1
        else:
            special += 1

    print("\nPassword Analysis:")
    if upper > 0:
        print("✔ Contains uppercase letter")
    else:
        print("✘ Missing uppercase letter")

    if lower > 0:
        print("✔ Contains lowercase letter")
    else:
        print("✘ Missing lowercase letter")

    if digit > 0:
        print("✔ Contains digit")
    else:
        print("✘ Missing digit")

    if special > 0:
        print("✔ Contains special character")
    else:
        print("✘ Missing special character")

    if upper > 0 and lower > 0 and digit > 0 and special > 0:
        print("\n Password is STRONG")
    else:
        print("\n⚠ Password is WEAK")

else:
    print("\n Password must be at least 8 characters long")
```

```python
# Online Shopping Cart

cart = {}   # dictionary to store items

while True:
    print("\n----- ONLINE SHOPPING CART -----")
    print("1. Add Product")
    print("2. Remove Product")
    print("3. View Cart")
    print("4. Total Bill")
    print("5. Exit")

    choice = int(input("Enter your choice: "))

    # ADD PRODUCT
    if choice == 1:
        product = input("Enter product name: ")
        price = int(input("Enter product price: "))
        quantity = int(input("Enter quantity: "))

        cart[product] = {"price": price, "quantity": quantity}
        print("Product added to cart!")

    # REMOVE PRODUCT
    elif choice == 2:
        product = input("Enter product name to remove: ")
        if product in cart:
            del cart[product]
            print("Product removed from cart.")
        else:
            print("Product not found in cart.")

    # VIEW CART
    elif choice == 3:
        if len(cart) == 0:
            print("Your cart is empty.")
        else:
            print("\nProduct | Price | Quantity")
            for i in cart:
                print(i, "|", cart[i]["price"], "|", cart[i]["quantity"])

    # TOTAL BILL
    elif choice == 4:
        total = 0
        for i in cart:
            total = total + (cart[i]["price"] * cart[i]["quantity"])
        print("Total Amount to Pay: ₹", total)

    # EXIT
    elif choice == 5:
        print("Thank you for shopping with us!")
        break

    else:
        print("Invalid choice. Try again.")
```

```python
# ATM System Using Multiple Accounts

accounts = {}          # store all accounts
current_account = None

while True:
    print("\n----- ATM MENU -----")
    print("1. Create Account")
    print("2. Login")
    print("3. Check Balance")
    print("4. Deposit Money")
    print("5. Withdraw Money")
    print("6. Logout")
    print("7. Exit")

    choice = int(input("Enter your choice: "))

    # CREATE ACCOUNT
    if choice == 1:
        acc_no = input("Enter Account Number: ")
        name = input("Enter Account Holder Name: ")
        pin = input("Set 4-digit PIN: ")
        balance = int(input("Enter Initial Balance: "))

        accounts[acc_no] = {
            "name": name,
            "pin": pin,
            "balance": balance
        }
        print("Account created successfully!")

    # LOGIN
    elif choice == 2:
        acc_no = input("Enter Account Number: ")
        pin = input("Enter PIN: ")

        if acc_no in accounts and accounts[acc_no]["pin"] == pin:
            current_account = acc_no
            print("Login successful. Welcome,", accounts[acc_no]["name"])
        else:
            print("Invalid account number or PIN.")

    # CHECK BALANCE
    elif choice == 3:
        if current_account is None:
            print("Please login first.")
        else:
            print("Current Balance: ₹", accounts[current_account]["balance"])

    # DEPOSIT
    elif choice == 4:
        if current_account is None:
            print("Please login first.")
        else:
            amount = int(input("Enter amount to deposit: "))
            accounts[current_account]["balance"] += amount
            print("Amount deposited successfully.")

    # WITHDRAW
    elif choice == 5:
        if current_account is None:
            print("Please login first.")
        else:
            amount = int(input("Enter amount to withdraw: "))
            if amount <= accounts[current_account]["balance"]:
                accounts[current_account]["balance"] -= amount
                print("Please collect your cash.")
            else:
                print("Insufficient balance.")

    # LOGOUT
    elif choice == 6:
        if current_account is not None:
            print("Logged out successfully.")
            current_account = None
        else:
            print("No account is logged in.")

    # EXIT
    elif choice == 7:
        print("Thank you for using ATM System.")
        break

    else:
        print("Invalid choice. Try again.")
```
