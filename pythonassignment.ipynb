```python
#1.Parking Lot Management: Track vehicle entry/exit, available spots, and compute parking fees
def parking_system():
    total_spots = int(input("Enter total parking spots: "))
    vehicles = {}

    while True:
        print("\n1. Vehicle Entry")
        print("2. Vehicle Exit")
        print("3. Show Available Spots")
        print("4. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            number = input("Enter vehicle number: ")
            hours = int(input("Enter parking hours: "))
            vehicles[number] = hours
            print("Vehicle parked")

        elif choice == "2":
            number = input("Enter vehicle number: ")
            if number in vehicles:
                hours = vehicles[number]
                fee = hours * 20
                print("Parking fee:", fee)
                del vehicles[number]
            else:
                print("Vehicle not found")

        elif choice == "3":
            print("Available spots:", total_spots - len(vehicles))

        elif choice == "4":
            break

parking_system()
```

```python
#2. to-Do List Manager with Categories: Add tasks, set deadlines, assign priority, and filter by project or tag.
def todo_manager():
    tasks = []

    while True:
        print("\n1. Add Task")
        print("2. Show Tasks")
        print("3. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            task = input("Enter task: ")
            category = input("Enter category: ")
            priority = input("Enter priority: ")
            tasks.append((task, category, priority))

        elif choice == "2":
            for t in tasks:
                print("Task:", t[0], "| Category:", t[1], "| Priority:", t[2])

        elif choice == "3":
            break

todo_manager()
```

```python
#3.Billing System for Retail Store: Scan products, apply discounts, generate bills, and record tinsactions.
def billing_system():
    cart = {}
    total = 0

    n = int(input("Enter number of products: "))

    for i in range(n):
        name = input("Enter product name: ")
        price = float(input("Enter price: "))
        qty = int(input("Enter quantity: "))
        cart[name] = price * qty
        total += price * qty

    discount = float(input("Enter discount %: "))
    final = total - (total * discount / 100)

    print("\nBill Summary")
    for item in cart:
        print(item, ":", cart[item])

    print("Total:", total)
    print("Final Amount:", final)

billing_system()
```

```python
#4.Movie Ticket Booking System: Check showtimes, book tickets, select seats, and print CI firmation.
def movie_booking():
    seats = 50
    movies = ["Avengers", "Pushpa", "KGF"]

    print("Available Movies:")
    for m in movies:
        print(m)

    movie = input("Choose movie: ")
    tickets = int(input("Enter number of tickets: "))

    if tickets <= seats:
        price = tickets * 150
        print("\nBooking Confirmed")
        print("Movie:", movie)
        print("Tickets:", tickets)
        print("Total Price:", price)
        seats -= tickets
    else:
        print("Not enough seats")

movie_booking()
```

```python
#5.College Result Management: Teachers enter marks, calculate GPA
def result_system():
    name = input("Enter student name: ")

    marks = {}
    subjects = ["Python", "Java", "DBMS"]

    for s in subjects:
        m = int(input("Enter marks for " + s + ": "))
        marks[s] = m

    total = sum(marks.values())
    avg = total / len(subjects)

    gpa = avg / 10

    print("\nStudent Result")
    print("Name:", name)
    print("Marks:", marks)
    print("Total:", total)
    print("Average:", avg)
    print("GPA:", gpa)

result_system()
```

```python
#1.Password Manager: Securely store, add, edit, and retrieve user passwords using encryption.
class PasswordManager:

    def __init__(self):
        self.passwords = {}

    def add_password(self):
        site = input("Enter website name: ")
        password = input("Enter password: ")
        self.passwords[site] = password
        print("Password saved")

    def view_password(self):
        site = input("Enter website name: ")
        if site in self.passwords:
            print("Password:", self.passwords[site])
        else:
            print("No password found")

    def show_all(self):
        print("Saved Accounts:")
        for site in self.passwords:
            print(site)

pm = PasswordManager()

pm.add_password()
pm.view_password()
pm.show_all()
```

```python
#2. P Freelancer Marketplace Simulation: Register freelancers and clients, assign projects, and cess payments
class FreelancerMarket:

    def __init__(self):
        self.freelancers = []
        self.projects = {}

    def register_freelancer(self):
        name = input("Enter freelancer name: ")
        self.freelancers.append(name)
        print("Freelancer registered")

    def assign_project(self):
        name = input("Enter freelancer name: ")
        project = input("Enter project name: ")
        payment = int(input("Enter payment amount: "))

        self.projects[name] = (project, payment)
        print("Project assigned")

    def show_projects(self):
        for f in self.projects:
            print("Freelancer:", f)
            print("Project:", self.projects[f][0])
            print("Payment:", self.projects[f][1])

fm = FreelancerMarket()

fm.register_freelancer()
fm.assign_project()
fm.show_projects()
```

```python
#3.Warehouse Automation System: Track goods movement, generate inventory reports, and forecast.
class Warehouse:

    def __init__(self):
        self.inventory = {}

    def add_goods(self):
        item = input("Enter item name: ")
        qty = int(input("Enter quantity: "))
        self.inventory[item] = qty

    def move_goods(self):
        item = input("Enter item name: ")
        move = int(input("Enter quantity moved: "))

        if item in self.inventory:
            self.inventory[item] -= move
            print("Goods moved")
        else:
            print("Item not found")

    def show_inventory(self):
        print("Inventory Report")
        for i in self.inventory:
            print(i, ":", self.inventory[i])

w = Warehouse()

w.add_goods()
w.move_goods()
w.show_inventory()
```

```python
#4. CRM (Customer Relationship Manager): Store customer info, manage communication logs, and track.
class CRM:

    def __init__(self):
        self.customers = {}

    def add_customer(self):
        name = input("Enter customer name: ")
        phone = input("Enter phone number: ")
        self.customers[name] = {"phone": phone, "logs": []}

    def add_log(self):
        name = input("Enter customer name: ")
        log = input("Enter communication note: ")

        if name in self.customers:
            self.customers[name]["logs"].append(log)

    def show_customer(self):
        for c in self.customers:
            print("Customer:", c)
            print("Phone:", self.customers[c]["phone"])
            print("Logs:", self.customers[c]["logs"])

crm = CRM()

crm.add_customer()
crm.add_log()
crm.show_customer()
```
